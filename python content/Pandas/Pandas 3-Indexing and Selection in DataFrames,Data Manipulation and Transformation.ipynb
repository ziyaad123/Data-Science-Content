## Boolean Indexing


we want to select all rows in the DataFrame where the Age is greater than or equal to 30. We can create a boolean mask using a comparison operation

```python
mask = df['Age'] >= 65
```

This creates a Series of booleans with the same length as the DataFrame, where each value is True if the corresponding value in the Age column is greater than or equal to 30, and False otherwise.

```python
# We can then use this mask to filter the DataFrame using the loc attribute:
df.loc[mask]
```

### We can also use boolean indexing to select specific columns based on a condition. 
### For example, to select all columns where the label contains the word "Type", 
### we can create a boolean mask using the str.contains method:

```python
mask = df.columns.str.contains('Type')
'''This creates a boolean mask with the same length as the number of columns, 
where each value is True if the corresponding column label contains the word "Type", and False otherwise.'''
```

```python
df.loc[:,mask]# we can see column name which has type keyowrd in it
```

### Boolean indexing is a powerful and flexible way to filter and select data in a DataFrame based on complex conditions.

LETS TAKE EXAMPLE IN THE DATA SET:

#### Example 1: Let's say we want to select all rows in the DataFrame where the Age is greater than or equal to 30 and the Income (USD) is greater than 5,000. We can create a boolean mask using comparison operations:

```python
mask=(df['Age'] >= 30) & (df['Income (USD)'] > 5000)
df.loc[mask]
```

The Mask creates a Series of booleans with the same length as the DataFrame, where each value is True if the corresponding row satisfies both conditions, and False otherwise.

This returns a new DataFrame containing only the rows where the mask is True.

#### EXAMPLE 2: Let's say we want to find out how many people have a credit score of 700 or above. We can do this using boolean indexing like this:

```python
mask = df['Credit Score'] >= 700
num_people = df[mask]['Name'].count()
```

# Data Manipulation and Transformation

### Reshaping:

1. **Pivoting**:
   - `pivot()`: Rotates data from rows into columns based on a key column.

2. **Melting**:
   - `melt()`: Converts data from wide format to long format.

3. **Stacking/Unstacking**:
   - `stack()`: Collapses DataFrame columns into a single column.
   - `unstack()`: Reverse operation of stacking.

4. **Reshaping with pivot_table**:
   - `pivot_table()`: Similar to `pivot`, but can handle duplicate entries by applying an aggregation function.

### Transforming:

1. **Grouping and Aggregation**:
   - `groupby()`: Groups data based on one or more keys.
   - `agg()`: Applies aggregate functions like sum, mean, etc.

2. **Resampling and Time-series Transformation**:
   - `resample()`: Resamples data over different time frequencies.
   
3. **Applying Functions**:
   - `apply()`: Applies custom functions to rows, columns, or entire DataFrame.

4. **Dropping Duplicates**:
   - `drop_duplicates()`: Removes duplicate rows based on a subset of columns or all columns.

5. **Renaming Columns**:
   - `rename()`: Renames columns to make them more readable or consistent.


# Reshaping:

## 1.Pivoting:
Suppose you want to reshape the data to have a separate column for each Type of Employment, with values being the Loan Amount Request (USD).

```python
df.pivot(index=['Customer ID', 'Name'], columns='Type of Employment', values='Loan Amount Request (USD)')
```

## 2.Melting:
If you want to convert the wide format of the dataset to a long format, where each row represents a single variable and its value, you can use melt().

```python
pd.melt(df, id_vars=['Customer ID', 'Name'], var_name='Variable', value_name='Value')
```

## 3.Stacking/Unstacking:
You can stack the Property ID, Property Age, Property Type, and Property Location columns into a single column.

```python
df.set_index(['Customer ID', 'Property ID'])[['Property Age', 'Property Type', 'Property Location']].stack().reset_index()
```

## 4.Reshaping with pivot_table:
Suppose you want to pivot the data to show the average Loan Amount Request (USD) for each combination of Location and Type of Employment, while handling potential duplicates.

```python
df.pivot_table(index='Location', columns='Type of Employment', values='Loan Amount Request (USD)', aggfunc='mean')
```

# Transforming:

## 1.Grouping and Aggregation:
You can group the data by Location and Type of Employment, and compute the mean Credit Score.

```python
df.groupby(['Location', 'Type of Employment'])['Credit Score'].mean().reset_index()
```

## 2.Applying Functions:
Suppose you want to create a new column Loan Amount Category based on the Loan Amount Request (USD) values.

```python
df['Loan Amount Category'] = df['Loan Amount Request (USD)'].apply(lambda x: 'High' if x > 100000 else 'Low')
df['Loan Amount Category'] 
```

## 3.Dropping Duplicates:
If you want to remove duplicate rows based on Customer ID and Property ID, you can use drop_duplicates().

```python
deduped_df = df.drop_duplicates(subset=['Customer ID', 'Property ID'])
deduped_df
```

## 4.Renaming Columns:
You can rename columns to make them more readable or consistent.

```python
renamed_df = df.rename(columns={'Income (USD)': 'Income', 'Loan Amount Request (USD)': 'Loan Amount'})
renamed_df.columns
```
