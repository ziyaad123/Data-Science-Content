- It is similar to a table or a spreadsheet, where data is organized in rows and columns.

# Data Inspection

- Data inspection is the process of examining a DataFrame to understand its structure, check for missing or inconsistent data, and learn about the data types of each column. 
- It provides insights into potential data issues and guides further data cleaning and analysis.

## Data Inspection Functions and their Meaning

1. **df.info()**: Provides information about the DataFrame, including the number of rows, columns, data types, and any missing data.
2. **df.head(n)**: Displays the first n rows of the DataFrame, useful for a quick overview.
3. **df.describe()**: Generates summary statistics for numerical columns, including count, mean, standard deviation, minimum, maximum, and quartiles.
4. **df.isnull().sum()**: Returns the count of missing values in each column, helping identify data quality issues.
5. **df.columns**: Lists the column names in the DataFrame.
6. **df.dtypes**: Shows the data types for each column, useful for detecting inconsistencies.

```python
# Check the structure and basic information of the DataFrame
df.info()
```

```python
# Display the first 5 rows to inspect the data
df.head(5)
```

```python
# Generate summary statistics for numerical columns
df.describe()
```

```python
# Check for missing values
print("Missing values per column:")
df.isnull().sum()
```

```python
#Lists the column names in the DataFrame.
df.columns
```

```python
#Shows the data types for each column
df.dtypes
```

# Data Exploration 

Data exploration is the process of analyzing the DataFrame to understand distributions, patterns, and relationships within the data. It helps to generate hypotheses and guide further analysis or data manipulation.

Data Exploration Functions and their Meaning
- **df["column"].value_counts()**: Returns the count of unique values in a categorical column, helping understand the distribution.
- **df["column"].unique()**: Lists all unique values in a column, useful for understanding diversity in the data.
- **df.corr()**: Computes the correlation matrix for numerical columns, indicating relationships between variables.

There are other techniques too which involves plots like histograms, heatmaps and scatterplot which we will learn and discuss in seaborn and matplotlib.

### .value_counts()

```python
# Get counts of unique values in the 'Gender' column
gender_counts = df["Gender"].value_counts()
print("Gender Counts:")
print(gender_counts)
```

Shows there are 10142 Females whereas 9827 males in the 'gender' columns

```python
# Get the count of unique values in the "Location" column
location_counts = df["Location"].value_counts()
print("Location Counts:")
print(location_counts)
```

Shows the number of location counts

### .unique()

```python
# Get the unique values in the "Gender" column
unique_genders = df["Gender"].unique()
print("Unique Genders:")
print(unique_genders)
```

Shows the number of unique values in the gender column

```python
# Get the unique values in the "Location" column
unique_locations = df["Location"].unique()
print("Unique Locations:")
print(unique_locations)
```

Shows the number of unique values in the 'location' column
