## Handling Missing Data

### Main Strategies:
1. **Remove**
   - Remove observations
   - Remove variables
   - Pairwise deletion

2. **Impute**
   - **Simple Impute** (Univariate)
     - Mean
     - Median
     - Mode
     - Constant value + add indicator for missingness
     - End of tails
   - **Multivariate Impute**
     - KNN impute
     - MICE (Multiple Imputation by Chained Equations)

***Univariate imputation** deals with missing data by looking at a single column at a time. It uses simple methods like filling in the missing values with the mean, median, or mode of that column.*

***Multivariate imputation**, on the other hand, considers multiple columns at once to estimate the missing values. It uses algorithms like KNN (K-Nearest Neighbors) or MICE (Multiple Imputation by Chained Equations) that take into account the relationships between different variables to predict the missing values more accurately.*

We would be starting form imputing values and with time would be reomoving rows if found not neccessary

## Imputing Null Values:
Replacing null values with meaningful estimates.
Strategies include mean, median, mode, or using more complex methods like regression or machine learning models.<br>
<b>fillna():</b> Fills null values with a specified value or method.

        <h1>Understanding Interpolation</h1>
        <p>Imagine you have dots on a piece of paper, but there are gaps between some of them. Interpolation helps you connect those dots with a smooth line or curve.</p>        
        <h2>When Do We Use Interpolation?</h2>
        <p>We use interpolation when we want to estimate values between known data points, especially in:</p>
        <ul>
            <li>Time Series Data</li>
            <li>Ordered Data (like distances along a path)</li>
            <li>Data with Smooth Trends</li>
        </ul>        
        <h2>Types of Interpolation Methods:</h2>
        <ul>
            <li><strong>Linear Interpolation:</strong> Connects dots with straight lines.</li>
            <li><strong>Polynomial Interpolation:</strong> Fits curves through dots.</li>
            <li><strong>Spline Interpolation:</strong> Uses smooth curves to fill gaps.</li>
            <li><strong>Nearest Neighbor Interpolation:</strong> Uses the value of the nearest dot to fill gaps.</li>
        </ul>        
        <h2>Why Use Interpolation?</h2>
        <p>Interpolation helps us complete missing data, maintain smoothness in our datasets, and make educated guesses about what values might be between our known points.</p>        
        <h2>Example:</h2>
        <p>If you have temperatures recorded every day but some days are missing, interpolation can estimate what the temperature might have been on those missing days based on the temperatures before and after.</p>        
        <p>Interpolation is a useful tool in data analysis and visualization, making our datasets more complete and helping us understand trends better.</p>

##  End of Distribution

"End of distribution" imputation is a method used to handle missing data where missing values are replaced with a value that lies at the end of the distribution of the variable.
- This value is often chosen based on percentiles, such as the 99th or 95th percentile, or based on a multiple of the standard deviation from the mean.

##### Advantages
- Preservation of variability
- Suitability for skewed distributions
- Minimal distortion of the data

##### Disadvantages
- Potential for introducing bias
- Sensitivity to outliers

#### Quartiles and End-of-Distribution
Quartiles divide a dataset into four equal parts:

- **Q1 (First Quartile)**: The 25th percentile; represents the median of the lower half of the data.
- **Q2 (Median)**: The 50th percentile; represents the middle value of the dataset.
- **Q3 (Third Quartile)**: The 75th percentile; represents the median of the upper half of the data.

These values can help identify the skewness of a distribution:
- If a distribution is **symmetrical**, the values of the quartiles will be evenly spaced.
- If a distribution is **skewed**, the quartiles will be unevenly spaced, indicating the direction of the skew.

#### Using Quartiles to Assess Skewness and Symmetry
To determine if a dataset is skewed or symmetrical, you can calculate the quartiles and visualize the data using box plots or other statistical methods.

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1  # Interquartile Range

# lower and upper whiskers
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

print("Lower Whisker:", lower_whisker)
print("Upper Whisker:", upper_whisker)
```
